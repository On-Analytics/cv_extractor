## CSV

```python
import pandas as pd
import ast

# Load your CSV
df = pd.read_csv('C:\\Users\Oscar\Downloads/extracted_data.csv')

# Step 1: Convert string columns to actual Python objects
df['items'] = df['items'].apply(ast.literal_eval)
df['metadata'] = df['metadata'].apply(ast.literal_eval)

# Step 2: Normalize 'metadata' dictionary into separate columns
metadata_df = pd.json_normalize(df['metadata'])
df = df.drop(columns='metadata').join(metadata_df)

# Step 3: Explode 'items' list into separate rows
df = df.explode('items').reset_index(drop=True)

# Step 4: Expand 'items' dict into its own columns
items_df = pd.json_normalize(df['items'])
df = df.drop(columns='items').join(items_df)

# Final tidy DataFrame
df.to_csv('tidy_invoices.csv', index=False)
```

```python
import pandas as pd
import ast

# Load Excel file
df = pd.read_excel('C:\\Users\Oscar\Downloads/extracted_data.xlsx')

# Step 1: Convert string columns to actual Python objects
df['items'] = df['items'].apply(ast.literal_eval)
df['metadata'] = df['metadata'].apply(ast.literal_eval)

# Step 2: Flatten metadata
metadata_df = pd.json_normalize(df['metadata'])
df = df.drop(columns='metadata').join(metadata_df)

# Step 3: Explode 'items'
df = df.explode('items').reset_index(drop=True)

# Step 4: Flatten 'items'
items_df = pd.json_normalize(df['items'])
df = df.drop(columns='items').join(items_df)

# Save to CSV or Excel
df.to_excel('tidy_invoices.xlsx', index=False)
```

## JSON

```python
import pandas as pd

# Load JSON file
df = pd.read_json('C:\\Users\Oscar\Downloads/extracted_data.json')

# Step 1: Flatten metadata
metadata_df = pd.json_normalize(df['metadata'])
df = df.drop(columns='metadata').join(metadata_df)

# Step 2: Explode items
df = df.explode('items').reset_index(drop=True)

# Step 3: Flatten items
items_df = pd.json_normalize(df['items'])
df = df.drop(columns='items').join(items_df)

# Save to Json
df.to_json('tidy_invoices.json', orient='records', indent=2)
```
